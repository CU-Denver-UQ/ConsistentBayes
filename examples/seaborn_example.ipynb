```python
import seaborn as sns
import numpy as np
import scipy.stats as sstats
```

```python
x_dist = sstats.norm(loc=0, scale=2)
y_dist = sstats.uniform(loc=0, scale=1)
```

```python
N = 10000
x_data = x_dist.rvs(N)
y_data = y_dist.rvs(N)
```

```python
colors_xkcd = sns.xkcd_palette(colors=["blue", "teal", "yellow"])
sns.jointplot(x=x_data, y=y_data, kind='kde', color=colors_xkcd[1], space=0.25, stat_func=None)
##
```
